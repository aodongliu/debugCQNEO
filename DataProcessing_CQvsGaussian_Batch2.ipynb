```python
# importing os module 
import os
#import the pandas library and aliasing as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
# This function reads the log file, extract key information and puts it in the dataframe
def fillDFGaussian(filename,df):
    temp = filename.split("/")[-1][:-4].split('_')

    index = filename.split("/")[-1][:-4].replace('_smt','').replace('_stw','')

    Alg = temp[1]

    # if the algorithm is simultaneous, we fill in the structure, basis information, and # of smt iterations
    if (index in df.index):
        df.loc[index]['Structure'] = temp[0]
        df.loc[index]['EBasis'] = temp[2]
        df.loc[index]['PBasis'] = temp[3]

        with open(filename, 'r') as f:
            for line in f.readlines():
                if 'NBasis=' in line:     
                    df.loc[index]['NEBasis'] = int(line.split()[1])
                    break

        with open(filename, 'r') as f:
            for line in f.readlines():
                if 'NEO NBasis=' in line:     
                    df.loc[index]['NPBasis'] = int(line.split()[2])
                    break

        E_Gau = 0.00
        with open(filename, 'r') as f:
            for line in f.readlines():
                if 'NEO-SCF Converged After Iteration #' in line:     
                    df.loc[index]['NIter-Gau'] = int(line.split()[5])
                if ('The Energy for This Iteration is:' in line):
                    E_Gau = float(line.split()[6])
        df.loc[index]['E-Gau'] =  E_Gau        
                   
```

```python
# This function reads the log file, extract key information and puts it in the dataframe
def fillDFCQ(filename,df):
    temp = filename.split("/")[-1][:-4].split('_')

    index = filename.split("/")[-1][:-4].replace('_CDIIS','').replace('_EDIIS','').replace('_CEDIIS','')

    DIISAlg = temp[-1]

    # if the algorithm is simultaneous, we fill in the structure, basis information, and # of smt iterations
    if (index in df.index):
        
        NIterColumnName = 'NIter-CQ-' + DIISAlg
        EColumnName = 'E-CQ-' + DIISAlg

        with open(filename, 'r') as f:
            for line in f.readlines():
                if 'SCF Completed: E(R-) =' in line:     
                    df.loc[index][NIterColumnName] = int(line.split()[7])
                    df.loc[index][EColumnName] = float(line.split()[4])
                    break

```

```python
# Main code starts here

gau_output_files = []
cq_output_files = []

gau_data_dir = '/Users/aodongliu/LiGroup/cq_neoscf_debug/data/gaussian_outputfiles'
cq_data_dir = '/Users/aodongliu/LiGroup/cq_neoscf_debug/data/cq_outputfiles/batch2'



for subdir, dirs, files in os.walk(gau_data_dir):
    for file in files:
        full_name = os.path.join(subdir, file)
        if file.endswith(".log"):
            gau_output_files.insert(0,full_name)
            
for subdir, dirs, files in os.walk(cq_data_dir):
    for file in files:
        full_name = os.path.join(subdir, file)
        if file.endswith(".out"):
            cq_output_files.insert(0,full_name) 
        
```

```python
structures = ['coh2','hcn','h2o']
elec_basis_pool = ['3-21g','6-311g','6-31g','def2svp','4-31g','def2tzvp','cc-pvdz','cc-pvtz','cc-pvqz']
prot_basis_pool = ['protsp','pb4d','pb4f1','pb5d','pb4f2','pb5f']
```

```python
Calculations = []
for structure in structures:
    for prot_basis in prot_basis_pool:                                          
        for elec_basis in elec_basis_pool: 
            elec_basis = elec_basis.replace('-', '').replace('(', '').replace(')', '').replace(',', '')
            prot_basis = prot_basis.replace('-', '').replace('(', '').replace(')', '').replace(',', '')
            temp = structure + '_' + elec_basis + '_' + prot_basis
            Calculations.append(temp)
```

```python
df = pd.DataFrame(columns=['Structure',
                          'EBasis',
                          'PBasis',
                          'NEBasis',
                          'NPBasis',
                          'NIter-Gau',
                          'NIter-CQ-CDIIS',
                          'NIter-CQ-EDIIS',
                          'NIter-CQ-CEDIIS',
                          'E-Gau',
                          'E-CQ-CDIIS',
                          'E-CQ-EDIIS',
                          'E-CQ-CEDIIS',
                          ],index= Calculations)
df
```

```python
for file in gau_output_files:
    fillDFGaussian(file,df)
df
```

```python
for file in cq_output_files:
    fillDFCQ(file,df)
df
```

```python
# Do some post processing to get the energy difference
df['E-Diff-C'] = df['E-Gau'] - df['E-CQ-CDIIS'] 
df['E-Diff-E'] = df['E-Gau'] - df['E-CQ-EDIIS'] 
df['E-Diff-CE'] = df['E-Gau'] - df['E-CQ-CEDIIS'] 
```

```python
df[['E-Diff-C-Grouped']] = df[['E-Diff-C']].fillna(value=-1)
df.loc[abs(df['E-Diff-C']) < 10e-7, 'E-Diff-C-Grouped'] = 0
df.loc[(abs(df['E-Diff-C']) > 10e-7) & (abs(df['E-Diff-C']) < 1), 'E-Diff-C-Grouped'] = 1
```

```python
df[['E-Diff-E-Grouped']] = df[['E-Diff-E']].fillna(value=-1)
df.loc[abs(df['E-Diff-E']) < 10e-7, 'E-Diff-E-Grouped'] = 0
df.loc[(abs(df['E-Diff-E']) > 10e-7) & (abs(df['E-Diff-E']) < 1), 'E-Diff-E-Grouped'] = 1
```

```python
df[['E-Diff-CE-Grouped']] = df[['E-Diff-CE']].fillna(value=-1)
df.loc[abs(df['E-Diff-CE']) < 10e-7, 'E-Diff-CE-Grouped'] = 0
df.loc[(abs(df['E-Diff-CE']) > 10e-7) & (abs(df['E-Diff-CE']) < 1), 'E-Diff-CE-Grouped'] = 1
```

```python
df.drop(['h2o_ccpvtz_protsp'], axis=0, inplace=True) 
df.drop(['h2o_ccpvqz_protsp'], axis=0, inplace=True) 
```

```python
def rand_jitter(arr):
    stdev = .01 * (max(arr) - min(arr))
    return arr + np.random.randn(len(arr)) * stdev
```

```python
fig, ax = plt.subplots(figsize=(12,8), dpi=600)


artist = ax.scatter(rand_jitter(df['NEBasis']),rand_jitter(df['NPBasis']),c=df['E-Diff-C-Grouped'] ,s=30)
ax.set_xlabel('Electronic Basis Size',size=20,weight='bold')
ax.set_ylabel('Protonic Basis Size',size=20,weight='bold')
ax.tick_params(axis='x', labelsize=20)
ax.tick_params(axis='y', labelsize=20)
cbar = plt.colorbar(artist)
cbar.set_label('0 for correct, 1 for wrong, -1 for not converged',size=20, weight='bold', labelpad=30, rotation=270)
cbar.set_ticks((-1,0,1))
plt.savefig('/Users/aodongliu/LiGroup/cq_neoscf_debug/plots/cq_vs_gau_cdiis_batch2.pdf', dpi=600)
plt.show()
```
