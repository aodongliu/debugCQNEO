```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cq_energy_nodiis  = np.loadtxt("cq_energy_nodiis.dat", dtype=None)
gau_energy_nodiis  = np.loadtxt("gau_energy_nodiis.dat", dtype=None)
```

```python
cq_energy_nodiis.size
```

```python
gau_energy_nodiis.size
```

```python
scf_iter = np.arange(1, cq_energy_nodiis.size+1,1)
```

```python
plt.style.use('default')
fig, ax = plt.subplots(1, 1,figsize =(12, 12),tight_layout=True)
ax.plot(scf_iter,cq_energy_nodiis,'o-',c='darkorange',ms=2,label="CQ")
ax.plot(scf_iter,gau_energy_nodiis,'o-',c='#9e66ab',ms=2,label="Gaussian")
plt.legend()
plt.savefig('./energy.pdf', dpi=3600)
```

```python
energy_diff = cq_energy_nodiis - gau_energy_nodiis
```

```python
plt.style.use('default')
fig, ax = plt.subplots(1, 1,figsize =(12, 12),tight_layout=True)
ax.plot(scf_iter,energy_diff,'o-',c='darkorange',ms=2,label="CQ")
ax.set_yscale("log")
plt.savefig('./energy_diff.pdf', dpi=3600)
```
